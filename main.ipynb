```python

### MY NEW CAFFE GHARGYY ELEVEN ####

print(" WELCOME TO MY CAFE GHARGYY ELEVEN: ☕")
print("Select Items:\n")

items = {
    "1": "Milk Tea 🍵",
    "2": "Black Tea",
    "3": "Coffee ☕",
    "4": "Latte Coffee",
    "5": "Green Tea 🍵"
}

# DisplayS Menu
for key, value in items.items():
    print(key + " - " + value)

selected_items = []

while True:
    choice = input("\n Select item number (1-5): ")

    if choice in items:
        print("You selected:", items[choice])
        selected_items.append(items[choice])
    else:
        print("Invalid selection! Please try again.")

    more = input("Do you want to add more items? (yes/no): ").lower()

    if more != "yes":
        break

print("\n Oh! Your Final Order:")
for item in selected_items:
    print("-", item)

print("\nTHANK YOU  PLEASE VISIT AGAIN! 💕")
```
